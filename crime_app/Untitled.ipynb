```python
import pandas as pd
from pandas import read_csv
```

```python
df = pd.read_csv("Criminal_incidents.csv")
```

```python
df.head()
```

```python
df["incidents"] = df["incidents"].replace(",", "")
df
```

```python
df.replace(',','', regex=True, inplace=True)
line_data = df.loc[:, ["year", "suburb", "incidents"]]

line_data[["incidents"]] = line_data[["incidents"]].apply(pd.to_numeric)
line_data
```

```python
line_group = line_data.groupby(["year", "suburb"])

suburb_total = line_group["incidents"].sum()

suburb_total
```

```python
new_df = pd.DataFrame(suburb_total)
new_df
```
